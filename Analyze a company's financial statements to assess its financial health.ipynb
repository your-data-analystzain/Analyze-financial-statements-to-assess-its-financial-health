```python
#Calculate the financial_ratios
def calculate_financial_ratios(df):
    df['Current Ratio'] = df['Total assets ($B)'] / df['Total liabilities ($B)']
    df['Return on Assets (ROA)'] = df['Earnings ($B)'] / df['Total assets ($B)']
    df['Return on Equity (ROE)'] = df['Earnings ($B)'] / df['Net assets ($B)']
    df['Debt-to-Equity Ratio'] = df['Total debt ($B)'] / df['Net assets ($B)']
    df['Asset Turnover Ratio'] = df['Revenue ($B)'] / df['Total assets ($B)']
    
    return df

# Calculate financial ratios
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df = calculate_financial_ratios(df)
print("Financial Ratios calculated:")
print(df[['Year', 'Current Ratio', 'Return on Assets (ROA)', 'Return on Equity (ROE)', 'Debt-to-Equity Ratio', 'Asset Turnover Ratio']])
```

```python
def plot_trends(df):
    plt.figure(figsize=(14, 10))
    
    # Subplot 1: Revenue and Earnings
    plt.subplot(2, 2, 1)
    plt.plot(df['Year'], df['Revenue ($B)'], label='Revenue ($B)')
    plt.plot(df['Year'], df['Earnings ($B)'], label='Earnings ($B)')
    plt.xlabel('Year')
    plt.ylabel('$B')
    plt.title('Revenue and Earnings Over Time')
    plt.legend()
    
    # Subplot 2: Market Cap and P/E Ratio
    plt.subplot(2, 2, 2)
    plt.plot(df['Year'], df['Market cap ($B)'], label='Market Cap ($B)')
    plt.plot(df['Year'], df['P/E ratio'], label='P/E Ratio')
    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title('Market Cap and P/E Ratio Over Time')
    plt.legend()
    
    # Subplot 3: Current Ratio
    plt.subplot(2, 2, 3)
    plt.plot(df['Year'], df['Current Ratio'], label='Current Ratio')
    plt.xlabel('Year')
    plt.ylabel('Ratio')
    plt.title('Current Ratio Over Time')
    plt.legend()
    
    # Subplot 4: ROA, ROE, and Debt-to-Equity Ratio
    plt.subplot(2, 2, 4)
    plt.plot(df['Year'], df['Return on Assets (ROA)'], label='ROA')
    plt.plot(df['Year'], df['Return on Equity (ROE)'], label='ROE')
    plt.plot(df['Year'], df['Debt-to-Equity Ratio'], label='Debt-to-Equity Ratio')
    plt.xlabel('Year')
    plt.ylabel('Ratio')
    plt.title('ROA, ROE, and Debt-to-Equity Ratio Over Time')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
plot_trends(df)
```
